```python
# Agora com esse novo data framre iremos realizar o plot do gráfico referente ao primeiro ponto de análise, em que iremos realizar a comparação da variação mensal do IPCA geral com o IPCA de cada setor que o compõem

# Converte a primeira coluna (datas) para o tipo datetime com o formato correto
ipca_variação_mensal_df['Data'] = pd.to_datetime(ipca_variação_mensal_df['Data'], format='%d/%m/%Y')

# Define as datas como índice do DataFrame
ipca_variação_mensal_df.set_index('Data', inplace=True)

# Obtém o número de colunas de dados
num_colunas = len(ipca_variação_mensal_df.columns)

# Cria um gráfico de linha para cada coluna de dados
plt.figure(figsize=(12, 6))

for i in range(num_colunas):
    plt.plot(ipca_variação_mensal_df.index, ipca_variação_mensal_df.iloc[:, i], label=ipca_variação_mensal_df.columns[i])

# Define rótulos e título
plt.xlabel('Data')
plt.ylabel('Variação(%)')
plt.title('Variação mensal do IPCA e de seus segmentos')

# Adiciona uma legenda
plt.legend()

# Rotaciona os rótulos do eixo x para melhorar a legibilidade
plt.xticks(rotation=45)

# Exibe o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Com a conclusão da nossa primeira análise iremos estudar o IPCA acumulado frente a taxa SELIC, esse que é um dos principais meio de controle disponíveis para o BACEM para que ele possa controla o a inflação

# Primeiramente iremos criar um data frame novo com os dados da Selic anualizada e do IPCA 12 meses, para que as informações estejam todas em um mesmo lugar

# Crie o DataFrame selic_x_ipca e adicione a coluna "Data" do ipca_df
selic_x_ipca = pd.DataFrame(ipca_df["Data"])

# Adicione a coluna "IPCA 12 meses" do ipca_df ao selic_x_ipca
selic_x_ipca["IPCA 12 meses"] = ipca_df["IPCA 12 meses"]

# Adicione a coluna "Selic" do selic_df ao selic_x_ipca
selic_x_ipca["Selic"] = selic_df["Selic"]

# O DataFrame selic_x_ipca agora contém as colunas desejadas de ambos os DataFrames
```

```python
# Agora que agrupamos nossos dados podemos realizar o plot do gráfico para compararmos os dois dados

# Define as datas como índice do DataFrame
selic_x_ipca.set_index('Data', inplace=True)

# Cria um gráfico de linha para as colunas "IPCA 12 meses" e "Selic"
plt.figure(figsize=(12, 6))

# Gráfico de linha para IPCA 12 meses
plt.plot(selic_x_ipca.index, selic_x_ipca['IPCA 12 meses'], label='IPCA 12 meses', marker='o', linestyle='-')

# Gráfico de linha para SELIC
plt.plot(selic_x_ipca.index, selic_x_ipca['Selic'], label='SELIC', marker='o', linestyle='-')

# Define rótulos e título
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Evolução do IPCA 12 meses e da SELIC')

# Adiciona uma legenda
plt.legend()

# Ajusta o espaçamento entre os rótulos do eixo x
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Exibe 10 rótulos no máximo

# Rotaciona os rótulos do eixo x para melhorar a legibilidade
plt.xticks(rotation=45)

# Exibe o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

Essa atitude foi capaz de fazer a inflação chegar ao valor final de 4,31% estando dentro das limitações perifericas da meta.

2) Em 2020 essas duas taxas foram muito afetadas pela pandemia. No meio do cenário pandemico, a atividade econômica desacerelou bastante, com isso ocorreu uma queda na taxa de inflação, como resposta o BACEM resolveu reduzir a taxa de juros, para tentar estimular o consumo. No entanto, no segundo semestre com o aumento do preço dos alimentos, a taxa de inflação voltou a subir, e o Banco Central optou por manter a selic em 1,90%, mas mantendo a cautela. Com isso a inflação finalizou o ano atingindo o valor de 4,52%.

3) Logo no início de 2023 a inflação ja começou a disparar, apresentando um aumneto muio relevando, muito impulsionada pelos fatores que foram citado na análise anterior. Com essa alta o Banco Central se preocupou em aumenta a taxa de juros, visando frear essa alta da inflação, no entanto, mesmo com esse aumento, a taxa continuava a fechando o ano em 10.06%, e a selic, por sua vez, também estava bem alto, na casa dos 8,76%, e mesmo assim, com esse alto patamar, a taxa de inflação continuava a subir.

4) Foi so no ano de 2022 que o BACEM conseguiu conter a taxa de inflção e induzi-la a reduzir de patamar. Mesmo que durante o ano o índice continuasse a crescer, atráves da alta taxa de juros estabelecida, ao final do ano a inflação a mostrava uma melhor significativa, conseguindo ser reduzida até o valor de 5,79%, e a selic, um dos principais responsáveis por tal efeito, fechou o ano com o valor mais alto dentre todos os períodos avaliados, no valor de 13,65%.

5) Para 2023 os planos do Banco Central se mativeram similares aos de 2022. Com o intuito de manter a queda da taxa de inflação continuaram com a sual política de uma elevada taxa Selic. É possível perceber que o momento de maior perigo ja passou, com as medidas tomadas pelo BACEM o IPCA mostrou melhoras significativas, tanto que nos últimos meses o processo de reduão da Selic ja iniciou, visto que não é nescessário que se matenha a taxa tao alta quanto antes para que seja possível alcançar a meta de 3,25% para a inflação esse ano, esse que foi projetada pelo COPOM.

Como uma idea geral do trabalho podemos perceber que a inflação flutuou bastante nesses últimos anos, e setores como alimentação e bebidas, trasnporte e educação (esse que em específico ja era esperado, dadas as características do setor) foram os que apresentaram variações mais significativas, e que a taxa selic se mostrou um mecanismo muito eficiente para que o Banco Central pudesse moldar a taxa de inflação. Essa capacidade da taxa de juros de influencia nesse cenário se da muito pela sua função de desencentivar o consumo e incentivar a poupança, uma que vez com taxas maiores o retorno dos inventimentos também sera maior, sendo melhor investir hoje e gastar no futuro. Isso se refelte na inflação porque ela representa o aumento dos preços e esse ponto é muito influenciado por quanto as pessoas estão gastando, quanto maior o consumo, maior a tendência do seu crescimento. Para os anos seguintes podemos esperar um afrouxamento da taxa de juros, visto que a inflação vem sendo controlada de forma efetiva. Portanto o consumo nos próximos anos deve voltar a crescer, acelerando a economia brasileira como um todo.
